```python
# Definindo as variáveis constantes
CATALOGO = "projeto_lakeflow"
SCHEMA = "bronze"
VOLUME = "dados_brutos"
NOME_ARQUIVO = "claims.csv"

# Montando o caminho dinamicamente
VOLUME_PATH = f"/Volumes/{CATALOGO}/{SCHEMA}/{VOLUME}/{NOME_ARQUIVO}"
NOME_TABELA = f"{CATALOGO}.{SCHEMA}.claims_bruto"

# Leitura do arquivo pelo Spark
df_bronze = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(VOLUME_PATH)

# Spark salva como tabela Delta nativa e otimizada
df_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(NOME_TABELA)

# Exibe o resultado na tela
display(df_bronze)
```
