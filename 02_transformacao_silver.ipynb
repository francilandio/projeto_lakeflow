```python
# Criando o banco de dados 'silver'
spark.sql("CREATE SCHEMA IF NOT EXISTS projeto_lakeflow.silver")

# Definindo de onde os dados vem e para onde vão
TABELA_BRONZE = "projeto_lakeflow.bronze.claims_bruto"
TABELA_SILVER = "projeto_lakeflow.silver.claims_limpo"

# Lendo os dados da tabela Bronze
df_bronze = spark.table(TABELA_BRONZE)

# Transformação Silver: Limpando duplicatas e remoção de valores nulos
df_silver = df_bronze.dropDuplicates().dropna()

# Salvando os dados validados na camada Silver
df_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(TABELA_SILVER)

# Exibe o resultado final
display(df_silver)
```
